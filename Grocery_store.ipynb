```python
class GroceryStore:
    def __init__(self):
        # Dictionary to store items and prices
        self.items = {
            "Rice": 50,
            "Sugar": 40,
            "Milk": 30,
            "Bread": 25,
            "Eggs": 20,
            "Salt": 15,
            "Pepper": 15,
            "Oil": 80,
        }

        #To display available items
    def display_items(self):
        print("\nAvailable Items:")
        for item, price in self.items.items():
            print(f"{item} - {price}")

    #To buy items
    def buy_item(self):
        try:
            item_name = input("\nEnter item name: ").title()

            if item_name not in self.items:
                raise ValueError("Item not available in store")

            quantity = int(input("Enter quantity: "))
            if quantity <= 0:
                raise ValueError("Quantity must be greater than zero")

            total_price = self.items[item_name] * quantity
            print(f"\nTotal price for {quantity} {item_name}(s): {total_price}")

        except ValueError as ve:
            print("Error:", ve)

        except Exception as e:
            print("Unexpected Error:", e)


# Main Program
store = GroceryStore()

while True:
    print("\n--- Grocery Store Menu ---")
    print("1. View Items")
    print("2. Buy Item")
    print("3. Exit")

    try:
        choice = int(input("Enter your choice: "))

        if choice == 1:
            store.display_items()
        elif choice == 2:
            store.buy_item()
        elif choice == 3:
            print("Thank you for shopping!")
            break
        else:
            print("Invalid choice. Please select 13.")

    except ValueError:
        print("Please enter a valid number.")
```
